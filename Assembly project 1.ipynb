## Описание данных


<b>Name</b> — название игры

<b>Platform</b> — платформа

<b>Year_of_Release</b> — год выпуска

<b>Genre</b> — жанр игры

<b>NA_sales</b> — продажи в Северной Америке (миллионы проданных копий)

<b>EU_sales</b> — продажи в Европе (миллионы проданных копий)

<b>JP_sales</b> — продажи в Японии (миллионы проданных копий)

<b>Other_sales</b> — продажи в других странах (миллионы проданных копий)

<b>Critic_Score</b> — оценка критиков (максимум 100)

<b>User_Score</b> — оценка пользователей (максимум 10)

<b>Rating</b> — рейтинг от организации ESRB (англ. Entertainment Software Rating 
Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

## Шаг 2. Подготовим данные

### Заменим названия столбцов (приведем к нижнему регистру)

```python
games.columns = games.columns.str.lower()
games.columns
```

```python
games.head()
```

### Обработаем пропуски

#### Исключим строки с неизвестным названием игры

```python
# таких всего 2
games = games.loc[~games['name'].isna()]
games.isna().sum()
```

Можно заметиить, что пустые строки с неизвестными жанрами также исчезли.

#### Удалим с пропущенными годами издания

```python
# 0.2% данных отбросим
games = games.loc[~games['year_of_release'].isna()]
games.info()
```

Интересно, а как можно было бы заполнить пропуски в годах?

#### Изучим пропуски в колонках с оценками и обработаем их

Посмотрим уникальные значения в колонке critic_score

```python
games['critic_score'].unique()
```

Заполним пропуски в колонке critic_score значением -10 (в дальнейшем оно преобразуется в заглушку -1) и приведем оценки к общему виду, как в user_score

```python
#games['critic_score'] = games['critic_score'].fillna(-10)
games['critic_score'] = games['critic_score']/10
games['critic_score'].unique()
```

Посмотрим уникальные значения в колонке user_score

```python
games['user_score'].unique()
```

Видим что есть уникальное значение tbd (To Be Determined - Будет определено), удалим его, после чего так же заполним пропуски заглушкой -1

```python
#заменим 'tbd' на nan
games.loc[games['user_score']=='tbd', 'user_score'] = np.nan
games['user_score'].unique()
```

```python
#games['user_score'] = games['user_score'].fillna(-1)
games['user_score'].unique()
```

Посмотрим на уникальные значения колонки rating

```python
games['rating'].unique()
```

Заполним пропуски Undefined

```python
games['rating'] = games['rating'].fillna('Undefined')
games['rating'].unique()
```

```python
games.isna().sum()
```

Больше нет пропусков

### Преобразуем типы данных

```python
games.info()
games['year_of_release'] = games['year_of_release'].astype('int')
games['user_score'] = games['user_score'].astype('float')
games['critic_score'] = games['critic_score'].astype('float')
```

```python
games.info()
```

### Посчитаем суммарные продажи во всех регионах и запишем их в отдельный столбец

```python
games['all_sales'] = games['na_sales'] + games['eu_sales'] + games['jp_sales'] + games['other_sales']
games.head()
```

## Шаг 3. Проведем исследовательский анализ данных

### Посмотрим сколько ирг выпускалось в разные годы

```python
#release_in_year = games.groupby('year_of_release')['name'].count()
release_in_year = games.pivot_table(index='year_of_release', values='name', aggfunc='count').sort_values(by='year_of_release', ascending = False)
(release_in_year.rename(columns = {'name' : 'count_games'}))
release_in_year.plot.barh(legend = False, figsize=(8, 12), title = 'Количество игр выпускаемых в год');
plt.xlabel('Количество игр, млн. проданных копий');
plt.ylabel('Годы');
```

Думаю, можно сказать, что наиболее интересными являются данные после 1994-1995, когда количество выпускаемых игр стало превышать 100

```python
games.query('year_of_release>1993')['name'].count()/games['name'].count()
```

Удалив данные с 1980 по 1993 год мы потеряем 2.2% данных, не критично.

```python
games = games.query('year_of_release>1993')
```

### Узнаем, за какой характерный срок появляются новые и исчезают старые платформы

```python
year = games.pivot_table(index='year_of_release',
                         columns='platform', 
                         values='all_sales', 
                         aggfunc='count')
sns.heatmap(year);
plt.xlabel('Платформы')
plt.ylabel('Годы жизни')
year[~year[:].isna()]=1
year.count().mean()
```

И так, можно сказать, что средняя продолжительность жизни плтаформы составляет около 7 лет.

### Возьмем данные за актуальный период. Данные за предыдущие годы отбросим.

_"Не учитывайте в работе данные за предыдущие годы."_

```python
games=games.query('2016>year_of_release>2012') # надеюсь я правильно понял задание
platform = games.groupby('platform')['all_sales'].sum() # year_of_realease
platform = platform.sort_values(ascending=False)
platform.plot(kind ='bar', figsize=(8,6));
```

Возьмем данные с платформ, которые имеют самые большие продажи за 3 года.

```python
best_platform = platform.index.to_list()
best_platform = best_platform[:10]
best_platform
```

```python
platform_in_year = games.query('platform in @best_platform')
platform_in_year_pivot = platform_in_year.pivot_table(index='year_of_release', 
                                                columns='platform', 
                                                values='all_sales', 
                                                aggfunc='sum')
platform_in_year_pivot.plot(grid=True, figsize=(15,7), style='-o')
#plt.xlim(1993,);
plt.xlabel('Годы');
plt.ylabel('Сумма продаж за год');
plt.title('Суммы продаж по годам, самых продаваемых платформ');
```

На графике видно, что игры на такие платформы как PS3 (PlayStation 3), Wii, WiiU, DS (Nintendo DS), PSP (Portative PlayStation), X360 (XBox360) перестали продаваться в последнее время.

Наиболее продаваемыми в последние годы были игры на платформах PS4, 3DS (Nintendo 3DS), PC (personal computer) (просто стабильная платформа по продажам последние 30 лет), XOne.

Все платформы падают в продажах, однако есть лидеры о которых я говорил ранее.

_Наиболее продаваемыми в последние годы были игры на платформах PS4, 3DS (Nintendo 3DS), PC (personal computer), XOne и WiiU._

```python
plt.figure(figsize=(15,15))
box = sns.boxplot(palette='Set1', data=platform_in_year_pivot);
plt.xlabel('Платформа');
plt.ylabel('Продажи');
```

Возьмем PS4 и XOne, т.к. это достаточно молодые платформы.

Не зависимо от того что медианные продажи на PC в 4 и 8 раз ниже чем на XOne и PS4 соответственно, данная платформа показывает стабильный доход с течением времени.

И до топ 5 добавим 3DS, WiiU, которым по ранее измеренным меркам жизни осталось пара лет.

```python
top_5_platform = ['PS4','XOne','PC','3DS','WiiU']
```

```python
platform_in_year = games.query('platform in @top_5_platform')
platform_in_year = platform_in_year.pivot_table(index='year_of_release', 
                                                columns='platform', 
                                                values='all_sales', 
                                                aggfunc='sum')
```

```python
plt.figure(figsize=(5,10))
box = sns.boxplot(palette='Set1', data=platform_in_year);
plt.xlabel('Платформа');
plt.ylabel('Продажи');
```

PS4 очень сильно выделяется на общем фоне, не смотря, что это одна из самых молодых платформ (вместе с XOne). Медианные продажи у XOne и 3DS в два раза ниже, чем у PS4, но при этом в два раза выше, чем у PC и WiiU.

### Посмотрим, как влияют на продажи внутри одной платформы отзывы покупателей и критиков.

```python
platform_best_user = games.query('platform == "PS4"')
platform_best_user = platform_best_user.loc[:, ['user_score','all_sales']]

display(platform_best_user.corr())
sns.set(rc={'figure.figsize':(10,5)}) #задаем размер графика
ax = sns.scatterplot(x="user_score", y="all_sales", data=platform_best_user, 
                     alpha=0.7,s=50)
ax.set(ylim=(0,10))
plt.xlabel('Оценка пользователей');
plt.ylabel('Сумма продаж');
```

Корреляция очень-очень слабая...

```python
platform_best_critic = games.query('platform == "PS4"')
platform_best_critic = platform_best_critic.loc[:, ['critic_score','all_sales']]
display(platform_best_critic.corr())
sns.set(rc={'figure.figsize':(10,5)}) #задаем размер графика
ax = sns.scatterplot(x="critic_score", y="all_sales", data=platform_best_critic, 
                     alpha=0.7,s=50)
ax.set(ylim=(0,10))
plt.xlabel('Оценка критиков');
plt.ylabel('Сумма продаж');
```

Зависимость продаж от оценок критиков в <s>2</s><b>*20*</b> раза выше чем от пользователей, но всё так же <s> мала.

```python
platform_best_user = games.query('platform == "XOne"')
platform_best_user = platform_best_user.loc[:, ['user_score','all_sales']]
display(platform_best_user.corr())
platform_best_critic = games.query('platform == "XOne"')
platform_best_critic = platform_best_critic.loc[:, ['critic_score','all_sales']]
display(platform_best_critic.corr())
```

```python
platform_best_user = games.query('platform == "PC"')
platform_best_user = platform_best_user.loc[:, ['user_score','all_sales']]
display(platform_best_user.corr())
platform_best_critic = games.query('platform == "PC"')
platform_best_critic = platform_best_critic.loc[:, ['critic_score','all_sales']]
display(platform_best_critic.corr())
```

Ситуация, в принципе, аналогична.

## Шаг 5. Проверим гипотезы

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Н0. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Н1. Средние пользовательские рейтинги платформ Xbox One и PC разные

```python
user_score_xone = games.dropna().query('platform == "XOne"') # данные с платформой XBox One
user_score_pc = games.dropna().query('platform == "PC"') # данные с платформой PC
print(len(user_score_xone))
print(len(user_score_pc))
```

```python
alpha = .05 # Зададим пороговое значение

results = st.ttest_ind(
    user_score_xone['user_score'], 
    user_score_pc['user_score'],
    ) 

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

```python
print('Средняя оценка пользователей платформы XBox One равна:', user_score_xone['user_score'].mean())
print('Средняя оценка пользователей платформы PC равна:', user_score_pc['user_score'].mean())
```

### Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Н0. Средние пользовательские рейтинги жанров Action и Sports сравнимые

Н1. Средние пользовательские рейтинги жанров Action и Sports разные

```python
user_score_action = games.dropna().query('genre == "Action"')
user_score_sports = games.dropna().query('genre == "Sports"')
print(len(user_score_action))
print(len(user_score_sports))
```

```python
alpha = .05 # Зададим пороговое значение

results = st.ttest_ind(
    user_score_action['user_score'], 
    user_score_sports['user_score'],
    equal_var = False) # т.к. объемы выборок отличаются в 3 раза

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

```python
print('Средняя оценка пользователей Action игр равна:', user_score_action['user_score'].mean())
print('Средняя оценка пользователей игр жанра Sports равна:', user_score_sports['user_score'].mean())
```

Альтернативные гипотезы были сформулированы в противовес нулевым. <b>Т.е. альтернативная гипотеза противоположна нулевой.</b>
